```python
car_crashes_df = sns.load_dataset('car_crashes')
car_crashes_df.head()
```

```python
sns.histplot(car_crashes_df.not_distracted, kde=True, color='red', linewidth=1);
```

```python
sns.jointplot(x=car_crashes_df.speeding, y=car_crashes_df.alcohol, kind='reg'); 
```

```python
sns.kdeplot(car_crashes_df.alcohol);
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips, hue='sex');
```

```python
sns.rugplot(tips.tip);
```

```python
sns.set_style('white') # white, whitegrid, dark, darkgrid, ticks
sns.set_context('paper', font_scale=1.3) # talk, poster

plt.figure(figsize=(8,4))

sns.jointplot(x=car_crashes_df.speeding, y=car_crashes_df.alcohol, kind='reg'); 

sns.despine(left=True, bottom=True, top=False, right=False);
```

```python
sns.barplot(x=tips.sex, y=tips.total_bill, estimator=np.mean)
```

```python
sns.barplot(x=tips.sex, y=tips.total_bill)
```

```python
sns.boxplot(x=tips.day, y=tips.total_bill, hue=tips.sex);
plt.legend(loc=1);
```

```python
sns.violinplot(x=tips.day, y=tips.total_bill, hue=tips.sex, split=True);
plt.legend(loc=1);
```

```python
sns.stripplot(x=tips.day, y=tips.total_bill, hue=tips.sex, jitter=True, dodge=True);
plt.legend(loc=0);
```

```python
sns.swarmplot(x=tips.day, y=tips.total_bill)
```

```python
sns.violinplot(x=tips.day, y=tips.total_bill)
sns.swarmplot(x=tips.day, y=tips.total_bill, color='white')
```

```python
car_crash_mx = car_crashes_df.corr()
sns.heatmap(car_crash_mx, annot=True)
```

```python
flights_df = sns.load_dataset('flights')
flights_df.head()
```

```python
flights_mx = flights_df.pivot_table(index='month', columns='year', values='passengers')
flights_mx
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights_mx, cmap="Blues", linecolor='white', linewidth=1)
```
